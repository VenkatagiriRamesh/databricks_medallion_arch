# 🏗️ Catalog & Schema Setup

**Purpose:** This notebook provisions the Unity Catalog infrastructure required by the
Medallion Architecture pipeline. It creates:

| Resource | Description |
|----------|-------------|
| `medallion_trial` catalog | Top-level namespace for all project data |
| `bronze` schema | Holds raw, unprocessed ingestion tables |
| `silver` schema | Holds cleaned, typed, and deduplicated tables |
| `gold` schema | Holds business-level aggregate tables |
| `bronze.raw_landing_zone` volume | External volume for landing raw CSV files |

> **Run this notebook once** before executing any of the ETL notebooks in `src/`.

---

```sql
-- =============================================================================
-- Step 1: Create the top-level Unity Catalog
-- =============================================================================
-- The catalog acts as the top-level namespace in Databricks Unity Catalog.
-- All schemas (bronze, silver, gold) will live under this catalog.
CREATE CATALOG IF NOT EXISTS medallion_trial;

-- Set the active catalog so subsequent commands don't need fully-qualified names.
USE CATALOG medallion_trial;

-- =============================================================================
-- Step 2: Create the three Medallion Architecture schemas
-- =============================================================================
-- Bronze: raw data, ingested as-is from source files
CREATE SCHEMA IF NOT EXISTS bronze;

-- Silver: cleaned, typed, and deduplicated data
CREATE SCHEMA IF NOT EXISTS silver;

-- Gold: business-ready aggregates and KPIs
CREATE SCHEMA IF NOT EXISTS gold;

-- =============================================================================
-- Step 3: Create a managed Volume for raw file landing
-- =============================================================================
-- Volumes are Unity Catalog's way of managing non-tabular files (CSVs, JSON, etc.).
-- Upload your raw Olist CSV files into this volume before running the Bronze notebook.
CREATE VOLUME IF NOT EXISTS bronze.raw_landing_zone;
```
