```sql
-- 1. Create the main Catalog for your project
CREATE CATALOG IF NOT EXISTS medallion_trial;

-- Set this catalog as the default for the rest of the commands
USE CATALOG medallion_trial;

-- 2. Create the three Medallion Schemas (Databases)
CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;

-- 3. Create a Volume inside the Bronze schema to act as your Raw Landing Zone
-- This is where your CSVs will live before they become Delta tables
CREATE VOLUME IF NOT EXISTS bronze.raw_landing_zone;
```
